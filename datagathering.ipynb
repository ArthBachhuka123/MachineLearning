```python
def language_check(language):
  if(language == "en"):
      return True
  else:
    return False
```

```python
ecom["Language"].apply(language_check).value_counts()
```

```python
def check_engineer(job):
  if "engineer" in job.lower():
    return True
  else:
    return False
```
